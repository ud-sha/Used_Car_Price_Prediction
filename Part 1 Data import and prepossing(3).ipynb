# Converting objects in integer

```python
#defined Empty string
empty_str = ""

#Function to convert string objects containing numerical details

#Method will only extract numerical value as strings and convert it into float
def Object_to_interger(x):
        value = empty_str.join(element for element in str(x) if (element.isdigit() or element =='.'))
        value = pd.to_numeric(value,errors = 'coerce')
        return value      
```

```python
df_copy['mileage'] = df_copy['mileage'].apply(Object_to_interger)
```

```python
df_copy['engine'] = df_copy['engine'].apply(Object_to_interger)
```

```python
df_copy['seats'] = df_copy['seats'].apply(Object_to_interger)
```

```python
df_copy['max_power'] = df_copy['max_power'].apply(Object_to_interger)
```

```python
df_copy['km_driven'] = df_copy['km_driven'].apply(Object_to_interger)
```

```python
df_copy.info()
```

```python
#Spliting Maker and Model using full_name
tem = df_copy["full_name"].str.upper()
df_copy['full_name'] = tem
new = df_copy["full_name"].str.split(" ", n = 3, expand = True)
  
# making separate first name column from new data frame
df_copy["Maker"]= new[0]
df_copy['Model']= new[1] + " " + new[2]
df_copy.drop('full_name', inplace=True, axis=1)
```

```python
df_copy.selling_price = df_copy.selling_price.str.replace('[*,,]', '')
df_copy[['selling_price','unit']] = df_copy.selling_price.str.split(expand=True)
df_copy['selling_price'] = df_copy['selling_price'].astype('float64', errors = 'raise')

df_copy.loc[df_copy.unit == "Lakh", 'selling_price'] = df_copy['selling_price']*100000.0
df_copy.loc[df_copy.unit == "Cr", 'selling_price'] = df_copy['selling_price']*10000000.0


df_copy=df_copy.drop(['unit'],axis=1)

df_copy.head()
```

```python
df_copy['new_price1']=df_copy['new-price'].str.lstrip('New Car (On-Road Price) : Rs.')
df_copy.new_price1 = df_copy.new_price1.str.replace('[*,,]', '')

df_copy[['new_price1','unit']] = df_copy.new_price1.str.split(" ",expand=True)

df_copy[['Min new price','Max new price']] = df_copy.new_price1.str.split("-",expand=True)
df_copy['Min new price'] = df_copy['Min new price'].str.replace('[A-Za-z]', '')
df_copy['Max new price'] = df_copy['Max new price'].str.replace('[A-Za-z]', '')

df_copy.drop(['new_price1'],axis=1)
df_copy.head(1)

# Changing datatype into float

df_copy['Max new price'] = df_copy['Max new price'].astype('float64', errors = 'raise')
df_copy['Min new price'] = df_copy['Min new price'].astype('float64', errors = 'raise')

# Converting cost price to appropriate units

df_copy.loc[df_copy.unit == "Lakh", 'Min new price'] = df_copy['Min new price']*100000.0
df_copy.loc[df_copy.unit == "Lakh", 'Max new price'] = df_copy['Max new price']*100000.0

df_copy.loc[df_copy.unit == "Cr", 'Min new price'] = df_copy['Min new price']*100000.0
df_copy.loc[df_copy.unit == "Cr", 'Max new price'] = df_copy['Max new price']*10000000.0

df_copy.drop(['unit','new_price1'],axis=1, inplace=True)
```

```python
#creating age column

# importing the date class datetime module
from datetime import date
# creating the date object of today's date
current_date = date.today()    
    
df_copy['Age'] = df_copy['year'].apply(lambda x : current_date.year - x)

df_copy.drop('year', inplace=True, axis=1)
```
